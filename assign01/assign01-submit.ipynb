```python
from collections import deque
```

```python
def print_tree(tree, level=0, label='.'): 
    print(' ' * (level*2) + label + ':' , tree.val)
    for child, lbl in zip([tree.left, tree.right], ['L', 'R']):  # do for all children 
        if child is not None:
            print_tree(child, level+1, lbl)
            
class BST():
    def __init__(self, val=None, parent=None, path_cost=0, action=None):
        self.val = val
        self.left = None
        self.right = None
        self.parent = parent
        self.path_cost = path_cost
        self.action = action
        
        
    def insert(self, val, action):
        if not self.val:
            self.val = val
            return
        
        if self.left is None:
            new_node = BST(val, parent=self, action=action)
            self.left = new_node
            return new_node
        else:
            self.left.insert(val, action)
            
        if self.right is None:
            # for right
            new_node = BST(val, parent=self, action=action)
            self.right = new_node
            return new_node
        else:
            self.right.insert(val, action)
```

```python
class Env:
    def __init__(self):
        self.steps = {'1':3,'2':4,'3':1,'4':3,'5':1,'6':3,'7':3,'8':3,'9':'G','10':2,'11':3,
                     '12':1,'13':2,'14':2,'15':3,'16':4,'17':2,'18':3,'19':3,'20':3,'21':4,'22':1,
                     '23':4,'24':3,'25':2}
        
        self.path_cost = {'1':1,'2':1,'3':1,'4':1,'5':1,'6':1,'7':1,'8':1,'9':'G','10':1,'11':1,
                     '12':1,'13':1,'14':1,'15':1,'16':1,'17':1,'18':1,'19':1,'20':1,'21':1,'22':1,
                     '23':1,'24':1,'25':1}
        
        self.movements = {'1':{'down':'6','right':'2'},
                          '2':{'left':'1','down':'7','right':'3',},
                          '3':{'left':'2','down':'8','right':'4'},
                          '4':{'left':'3','down':'9','right':'5'},
                          '5':{'left':'4','down':'10'},
                          '6':{'down':'11','right':'7','up':'1'},
                          '7':{'down':'12','left':'6','up':'2','right':'8'},
                          '8':{'up':'3','right':'9','left':'7','down':'13'},
                          '9':{'down':'14','up':'4','left':'8','right':'10'},
                          '10':{'up':'5','down':'15','left':'9'},
                          '11':{'up':'6','down':'16','right':'12'},
                          '12':{'up':'7','down':'17','right':'13','left':'11'},
                          '13':{'up':'8','down':'18','right':'14','left':'12'},
                          '14':{'up':'9','down':'19','right':'15','left':'13'},
                          '15':{'up':'10','down':'20','left':'14'},
                          '16':{'up':'11','down':'21','right':'17'},
                          '17':{'up':'12','down':'22','right':'18','left':'16'},
                          '18':{'up':'13','down':'23','right':'19','left':'17'},
                          '19':{'up':'14','down':'24','right':'20','left':'18'},
                          '20':{'up':'15','down':'25','left':'19'},
                          '21':{'up':'16','right':'22'},
                          '22':{'up':'17','right':'23','left':'21'},
                          '23':{'up':'18','right':'24','left':'22'},
                          '24':{'up':'19','right':'25','left':'23'},
                          '25':{'up':'20','left':'24'}}
        
    def initial_state(self,tree):
        if tree == None:
            return '1'

        steps_allowed = []
        steps_allowed.append(self.steps['1'],self.steps['2'],self.steps['3'],self.steps['4'],self.steps['5'])
        steps_allowed.append(self.steps['6'],self.steps['7'],self.steps['8'],self.steps['9'],self.steps['10'])
        steps_allowed.append(self.steps['11'],self.steps['12'],self.steps['13'],self.steps['14'],self.steps['15'])
        steps_allowed.append(self.steps['16'],self.steps['17'],self.steps['18'],self.steps['19'],self.steps['20'])
        steps_allowed.append(self.steps['21'],self.steps['22'],self.steps['23'],self.steps['24'],self.steps['25'])

        for i in steps_allowed:
            string = ""
            for j in i:
                string = string+" "+str(j)
            print(string)
                
    def goal_test(self, node):
        if self.steps[node.val] == 'G':
            return True
        else:
            return False
        
    def actions(self,node):
        actions = []
        
        for action in self.movements[node.val].keys():
            actions.append(action)
            
        return actions
```

```python
def bfs(Env):
    
    state = Env.initial_state(None)
    node = BST(state, parent=None, path_cost=0, action=None)
    
    frontier = deque()
    explored = []
    
    if Env.goal_test(node):
        return (node.val, path_cost)
    
    frontier.append(node)
    
    while frontier:
        
        # popleft: to get 0th index element from queue (builtin in 'deque' library)
        node = frontier.popleft()
        print('agent is currently at node',node.val)
        
        explored.append(node)
        
        for action in Env.actions(node):
            child = child_node(Env, node, action)
            print('child', child.val, 'is created')
            
            if child not in frontier or child not in explored:
                print(child.val,'not in frontier and explored')
                
                if Env.goal_test(child):
                    return child
                
                frontier.append(child)
                
        
BST.bfs = bfs
```

```python
def child_node(Env, parent, action):
    path_cost = 0
    
    if action == 'down':
        dwn = Env.movements[parent.val][action]
        if Env.steps[dwn] != 'G':
            path_cost = Env.path_cost[parent.val] + Env.steps[dwn]
#         child = BST(Env.movements[parent][action], parent, path_cost, action=None)
        child = parent.insert(Env.movements[parent.val][action], action)
    
        return child
    
    if action == 'up':
        up = Env.movements[parent.val][action]
        if Env.steps[up] != 'G':
            path_cost = Env.path_cost[parent.val] + Env.steps[up]
#         child = BST(Env.movements[parent.val][action], parent, path_cost, action=None)
        child = parent.insert(Env.movements[parent.val][action], action)
    
        return child
    
    if action == 'right':
        rgt = Env.movements[parent.val][action]
        if Env.steps[rgt] != 'G':
            path_cost = Env.path_cost[parent.val] + Env.steps[rgt]
#         child = BST(Env.movements[parent.val][action], parent, path_cost, action=None)
        child = parent.insert(Env.movements[parent.val][action], action)
    
        return child
    
    if action == 'left':
        lft = Env.movements[parent.val][action]
        if Env.steps[lft] != 'G':
            path_cost = Env.path_cost[parent.val] + Env.steps[lft]
#         child = BST(Env.movements[parent.val][action], parent, path_cost, action=None)
        child = parent.insert(Env.movements[parent.val][action], action)
    
        return child
```
